```python
from pathlib import Path
import os
# path of the video file to be cropped
path = r"C:\Users\baido\Desktop\DifE\drive\Simba\video editing with python"

output_video = r"C:\Users\baido\Desktop\DifE\drive\Simba\video editing with python\cropped_video"

w=640
h=480

for filename in os.listdir(path):
    if ".mp4" in filename:
        tt=os.path.join(path, filename)
        baseName=filename[0:20]
        lastName=filename[36:]
        name1=filename[21:26]
        name2=filename[31:36]
        fname1=name1+ "_" +baseName+lastName
        fname2=name2+"_" + baseName+lastName
        #t1=len(filename)
        #rint(fname1)
        #rint(fname2)       
        #ilename1=filename[:]
        clip = mp.VideoFileClip(tt)
        clipsizet=clip.size
        #rint(clipsizet)
        if(w and h) in clipsizet:
            cropped_clip1 = crop(clip, x1=50,  x2=330, y1=100, y2=380)
            cropped_clip2 = crop(clip, x1=325, width=280, y1=100, y2=380)
            out1=os.path.join(output_video, fname1)
            out2=os.path.join(output_video, fname2)
            cropped_clip1.write_videofile(out1,codec="libx264")
            cropped_clip2.write_videofile(out2,codec="libx264")
```
